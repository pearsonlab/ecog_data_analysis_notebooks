```python
import numpy as np
import matplotlib.pyplot as plt

import mne
from mne.time_frequency import tfr_multitaper, tfr_morlet
from mne.stats import permutation_cluster_test
import ecogtools
```

```python
patient_num = "2002"

channels = ['RTG22', 'RTG23', 'RTG24', 'RTG30', 'RTG31', 'RTG32', 'RTG38', 'RTG39', 'RTG40', 'RTG46', 'RTG47', 'RTG48']
```

```python
data = ecogtools.ToM_Localizer(patient_num)

# Get the times for task from behavioral data
tmin = int(data.trig_and_behav.loc[0, 'trigger_time']-10)
tmax = int(data.trig_and_behav.loc[59, 'trigger_time']+10)

# Get only physiology data that we actually want to look at for task.
data.phys.crop(tmin=tmin, tmax=tmax)
```

```python
# Load data and pick only channels that we care about.
data.phys.load_data()
data.phys.pick_channels(channels)
```

```python
mne.set_log_level(verbose=False)
```

```python
tmin, tmax = -3, 1.5

for i, chan in zip(range(len(channels)), channels):
    print(chan)
    ch_name = chan # restrict example to one channel

    # Load condition 1
    event_id = 16
    epochs_condition_1 = mne.Epochs(data.phys, data.events, event_id, tmin, tmax,
                                    baseline=None,
                                    preload=True)

    epochs_condition_1.pick_channels([ch_name])

    # Load condition 2
    event_id = 17
    epochs_condition_2 = mne.Epochs(data.phys, data.events, event_id, tmin, tmax,
                                    baseline=None,
                                    preload=True)
    epochs_condition_2.pick_channels([ch_name])


    frequencies = np.arange(80, 150, 3)  # define frequencies of interest
    n_cycles = frequencies/2.
    time_bandwidth = 8

    tfr_epochs_1 = tfr_multitaper(epochs_condition_1, frequencies, time_bandwidth=time_bandwidth,
                              n_cycles=n_cycles,
                              return_itc=False, average=False)

    tfr_epochs_2 = tfr_multitaper(epochs_condition_2, frequencies, time_bandwidth=time_bandwidth,
                              n_cycles=n_cycles,
                              return_itc=False, average=False)

    tfr_epochs_1.apply_baseline(mode='ratio', baseline=(0, .5))
    tfr_epochs_2.apply_baseline(mode='ratio', baseline=(0, .5))
    tfr_epochs_1.crop(-1, .5)
    tfr_epochs_2.crop(-1, .5)

    epochs_power_1 = tfr_epochs_1.data[:, 0, :, :]  # only 1 channel as 3D matrix
    epochs_power_2 = tfr_epochs_2.data[:, 0, :, :]  # only 1 channel as 3D matrix

    tfr_epochs_1.average().plot([0], dB=True, vmin=-5, vmax=5);

    tfr_epochs_2.average().plot([0], dB=True, vmin=-5, vmax=5);

    """   
    Compute statistic
    -----------------

    """

    threshold = 3.0
    T_obs, clusters, cluster_p_values, H0 = \
        permutation_cluster_test([epochs_power_1, epochs_power_2],
                                 n_permutations=100, threshold=threshold, tail=0)
    """
    View time-frequency plots
    """


    times = 1e3 * tfr_epochs_1.times  # change unit to ms

    plt.figure()
    plt.subplots_adjust(0.12, 0.08, 0.96, 0.94, 0.2, 0.43)

    plt.subplot(1, 1, 1)
    # Create new stats image with only significant clusters
    T_obs_plot = np.nan * np.ones_like(T_obs)
    for c, p_val in zip(clusters, cluster_p_values):
        if p_val <= 0.05:
            T_obs_plot[c] = T_obs[c]

    plt.imshow(T_obs,
               extent=[times[0], times[-1], frequencies[0], frequencies[-1]],
               aspect='auto', origin='lower', cmap='gray')
    plt.imshow(T_obs_plot,
               extent=[times[0], times[-1], frequencies[0], frequencies[-1]],
               aspect='auto', origin='lower', cmap='RdBu_r')

    plt.colorbar()
    plt.xlabel('Time (ms)')
    plt.ylabel('Frequency (Hz)')
    plt.title('Induced power (%s)' % ch_name)

    plt.show()
    print('\n')
```
